```python
import cv2
import easyocr
import json

# 이미지 경로
image_path = '/Users/402-07/train/IMG_OCR_53_4PO_09451.png'

# OpenCV를 사용하여 이미지 읽기
image = cv2.imread(image_path)

# EasyOCR을 초기화하고 사용할 언어를 설정
reader = easyocr.Reader(['en', 'ko'], gpu=False)  # 필요한 언어 코드로 변경

# 이미지에서 텍스트 추출
result = reader.readtext(image)

# JSON 객체 초기화
columns = []

# 추출된 텍스트 출력
for detection in result:
    text = detection[1]
    bounding_box = detection[0]
    
    # "상호" 컬럼 처리
    if "상호" in text:
        x_min, y_min, x_max, y_max = bounding_box[812][653], bounding_box[812][726], bounding_box[938][653], bounding_box[938][726]

        # JSON 객체에 컬럼 추가
        column = {
            "name": "상호",
            # 박스 왼쪽 위부터 오른쪽 아래 좌표 추가
            "coordinates": {
                "x_min": 812,
                "y_min": 726,
                "x_max": 938,
                "y_max": 653
            }
        }
        columns.append(column)

        # 텍스트 영역을 사각형으로 감싸기 (선택 사항)
        cv2.rectangle(image, (812, 726), (938, 653), (0, 255, 0), 2)

        # 텍스트 출력
        cv2.putText(image, text, (int(x_min), int(y_min) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
#     # "이름" 컬럼 처리
#     elif "성명" in text:
#         # 인덱스에 주의: 예시로 인덱스 0과 1을 사용
#         x_min, y_min, x_max, y_max = bounding_box[0][0], bounding_box[0][1], bounding_box[1][0], bounding_box[1][1]

#         # JSON 객체에 컬럼 추가
#         column = {
#             "name": "성명",
#             "coordinates": {
#                 "x_min": int(x_min),
#                 "y_min": int(y_min),
#                 "x_max": int(x_max),
#                 "y_max": int(y_max)
#             }
#         }
#         columns.append(column)

#         # 텍스트 영역을 사각형으로 감싸기 (선택 사항)
#         cv2.rectangle(image, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0, 255, 0), 2)

#         # 텍스트 출력
#         cv2.putText(image, text, (int(x_min), int(y_min) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# # JSON 객체를 파일에 저장
# with open('columns.json', 'w') as json_file:
#     json.dump(columns, json_file)

# # 결과 이미지 저장 또는 표시
# cv2.imwrite('output_image.jpg', image)
# cv2.imshow('Image with Text Detection', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
